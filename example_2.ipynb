```python
from data_cleaner import DataCleaner

cleaner = DataCleaner()

print("=== String Cleaning ===")
text = "  Hello   World!  @#$  "
print(f"Original: '{text}'")
print(f"Cleaned: '{cleaner.clean_string(text, strip=True, remove_extra_spaces=True, remove_special_chars=True)}'")

# Numeric cleaning
print("\n=== Numeric Cleaning ===")
numeric_tests = ["$1,234.56", "5kg", "-10", "invalid"]
for test in numeric_tests:
    print(f"'{test}' -> {cleaner.clean_numeric(test, default=0)}")

# Email cleaning
print("\n=== Email Cleaning ===")
emails = ["user@example.com", "invalid.email@", "USER@EXAMPLE.COM"]
for email in emails:
    print(f"'{email}' -> {cleaner.clean_email(email)}")

# Phone cleaning
print("\n=== Phone Cleaning ===")
phones = ["(555) 123-4567", "5551234567", "invalid"]
for phone in phones:
    print(f"'{phone}' -> {cleaner.clean_phone(phone)}")

# Date cleaning
print("\n=== Date Cleaning ===")
dates = ["2023-12-25", "25/12/2023", "Dec 25, 2023"]
for date in dates:
    print(f"'{date}' -> {cleaner.clean_date(date)}")

# List cleaning
print("\n=== List Cleaning ===")
test_list = [1, 2, 2, 3, "", None, 4, 4, 5]
print(f"Original: {test_list}")
print(f"Cleaned: {cleaner.clean_list(test_list)}")

# Outlier removal
print("\n=== Outlier Removal ===")
data = [1, 2, 3, 4, 5, 100, 102, 3, 4, 5]
print(f"Original: {data}")
print(f"Without outliers: {cleaner.remove_outliers_iqr(data)}")

# Text standardization
print("\n=== Text Standardization ===")
sample_text = "The quick brown foxes are jumping over the lazy dogs!"
print(f"Original: {sample_text}")
print(f"Standardized: {cleaner.standardize_text(sample_text, remove_stopwords=True, stem=True)}")
```
